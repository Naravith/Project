```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('Dataset\GOOG.csv',date_parser = True)
data.tail()
```

```python
data_training =  data[ data['Date'] < '2019-01-01' ].copy()
data_training
```

```python
data_test =  data[ data['Date']>= '2019-01-01' ].copy()
data_test
```

```python
# We are going to predict 'OPEN'

training_data = data_training.drop(['Date','Adj Close'], axis=1)
training_data.head()
```

```python
scaler = MinMaxScaler()
training_data = scaler.fit_transform(training_data)
training_data
```

```python
X_train = []
y_train = []
```

```python
training_data.shape[0]
```

```python
# 60 days per iteration

for i in range(60,training_data.shape[0]):
    X_train.append(training_data[i-60:i])
    y_train.append(training_data[i, 0]) #column à¸—à¸µà¹ˆ 0 = OPEN
```

```python
#Convert to numpy array to be input of NN
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_train.shape, y_train.shape

# 3556 row, chunk of 60/steps, 5 features
```

# Building LSTM

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
# Name our model { regressior }
regressior = Sequential()

regressior.add( LSTM(units=60, activation='relu', return_sequences= True, input_shape=(X_train.shape[1],5 )) )
regressior.add(Dropout(0.2))

regressior.add( LSTM(units=60, activation='relu', return_sequences= True) )
regressior.add(Dropout(0.2))

regressior.add( LSTM(units=80, activation='relu', return_sequences= True) )
regressior.add(Dropout(0.2))

#Final Layer of LSTM
regressior.add( LSTM(units=120, activation='relu') )
regressior.add(Dropout(0.2))

#output layer
regressior.add(Dense(units=1))
```

```python
regressior.summary()
```

```python
regressior.compile(optimizer='adam', loss='mean_squared_error')
```

```python
regressior.fit(X_train, y_train, epochs=10, batch_size=32)
```

### Prepare Test Data 

```python
past_60_days = data_training.tail(60)
```

```python
df = past_60_days.append(data_test, ignore_index=True)
df = df.drop(['Date','Adj Close'], axis=1)
df
```

```python
inputs = scaler.transform(df) #no need to fit the test data, transform only
inputs
```

```python
X_test, y_test = [],[]
```

```python
for i in range(60,inputs.shape[0]):
    X_test.append(inputs[i-60:i])
    y_test.append(inputs[i,0])
```

```python
X_test = np.array(X_test)
y_test = np.array(y_test)

X_test.shape, y_test.shape
```

```python
y_pred = regressior.predict(X_test)
y_pred
```

```python
# We have to inverse to scale back to normal

scaler.scale_
```

```python
# 8.18605127e-04 = scale for first column (OPEN)
scale = 1/8.18605127e-04
```

```python
y_pred *= scale
y_test *= scale
```

```python
y_pred
```

### Visualization 

```python
plt.figure(figsize =(14,5))
plt.plot(y_test, color='red', label='Real stock price')
plt.plot(y_pred, color='blue', label='Predicted stock price')

plt.title('Google stock price prediction')
plt.xlabel('Time')
plt.ylabel('Google Stock Price')

plt.legend()
plt.show()
```
